```python
#Scraping Wikipedia for total nobel prizes by country details

wikipedia_url="https://en.wikipedia.org/wiki/List_of_Nobel_laureates_by_country#Summary"

wikipedia_summary=pd.read_html(wikipedia_url)

#Creating a dataframe and load with content from table 3
summary_df=wikipedia_summary[2]

summary_df=summary_df.rename(columns={
                                    'Name': 'Country',
                                    'Number of Nobel laureates(Number of Nobel Prizes)' : 'Number of Nobel Prizes'
})
```

```python
np_sql="""select np_dtl.year as "Year",np_dtl.country as "Country",np_dtl.category as "Category",np_dtl.name as "Name",
                    np_dtl.gender as "Gender",count(*) as "count" FROM
                    nobelprize_details np_dtl
                    group by  np_dtl.year,np_dtl.country,np_dtl.category, np_dtl.name,np_dtl.gender"""
np_count_df=pd.read_sql(np_sql, connection)

np_count_df.head()
```

```python
nobelprize_women_df.head()
```

```python
#np_gender_group=np_count_df.groupby(['Category','Gender']).size().reset_index(name="Gender and Category count")

np_man_group=np_count_df.loc[np_count_df['Gender']=='Man'].groupby(['Category','Gender']).size().reset_index(name="Gender and Category count")

np_women_group=np_count_df.loc[np_count_df['Gender']=='Woman'].groupby(['Category','Gender']).size().reset_index(name="Gender and Category count")

# x_axis=
```

```python
combine_np_data_df=pd.merge(np_man_group,np_women_group, how='outer', on='Category')

combine_np_data_df=combine_np_data_df.rename(columns={
                                               "Gender and Category count_x": "Men_NP_count",
                                               "Gender and Category count_y": "Women_NP_count",
                                                "Gender_x" :"Men",
                                                "Gender_y" :"Women"
})

combine_np_data_df
```

```python
# np_gender_group.plot(kind="bar",figsize=(12,3))
# plt.show

combine_np_data_df.plot(x='Category' , y= ['Men_NP_count', 'Women_NP_count'], kind='bar',stacked=True)
plt.title("Nobel Prize by Category and Gender")
plt.ylabel("No of Nobel Prizes")
plt.grid()
plt.savefig("NobelPrize-CategoryGender.png")
plt.show
```

```python
np_women_b41980 =np_count_df.loc[(np_count_df['Gender']=='Woman') & (np_count_df['Year']<1979)].groupby(['Category','Gender']).size().reset_index(name="Category count")

np_women_after1980 =np_count_df.loc[(np_count_df['Gender']=='Woman') & (np_count_df['Year']>1979)].groupby(['Category','Gender']).size().reset_index(name="Category count")


combine_women_year_df=pd.merge(np_women_b41980,np_women_after1980, how='outer', on='Category')

combine_women_year_df=combine_women_year_df.drop("Gender_x",axis=1)
combine_women_year_df=combine_women_year_df.drop("Gender_y",axis=1)
combine_women_year_df.fillna(0, inplace=True)
combine_women_year_df=combine_women_year_df.rename(columns={
                                               "Category count_x": "Women_NP_b41980",
                                               "Category count_y": "Women_NP_after_1980"
})

combine_women_year_df
```

```python
combine_women_year_df.plot(x='Category' , y= ['Women_NP_b41980', 'Women_NP_after_1980'], kind='bar',stacked=True)
plt.title("Women: Nobel Prize comparison in last 20 years Vs 1901-1979")
plt.ylabel("No of Nobel Prizes")
plt.grid()
plt.savefig("NobelPrize-WomenProgress.png")
plt.show
```
